```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
img = plt.imread('lamem.jpg')

plt.figure(figsize=(10,10))
plt.imshow(img)

plt.show()
```

```python
def kmeanR(nRodadas, img, K):
    
    Z = img.reshape((-1,3))
    Z = np.float32(Z)

    criteria = (cv.TERM_CRITERIA_EPS + cv.TERM_CRITERIA_MAX_ITER, nRodadas, 0.0001)
    ret,label,center=cv.kmeans(Z,K,None,criteria,10,cv.KMEANS_RANDOM_CENTERS)

    center = np.uint8(center)
    res = center[label.flatten()]
    image = res.reshape((img.shape))
    
    return image
```

Executando o kmeans várias vezes. Número de K centróides igual a 8. 10 épocas de treino

```python
for i in range(0, 10):
    img_k = kmeanR(nRodadas=1, img=img, K=8)
    cv.imwrite('result'+str(i)+'.jpg', img_k)
```

```python
f, ax = plt.subplots(5, 2, figsize=(20, 20))

count = 0

for i in range(5):
    for j in range(2):
        img = plt.imread('result'+str(count)+'.jpg')
        ax[i][j].imshow(img)
        count += 1

plt.show()
```
