Lesson 1: This notebook convert image to gray, get 2 points and apply negative filter

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Read image as gray

```python
img = cv2.imread("biel.png", cv2.COLOR_RGB2GRAY)
```

Dimensions

```python
img.shape
```

```python
P1x = int(input("Enter with P1x:"))
P1y = int(input("Enter with P1y:"))
```

```python
P2x = int(input("Enter with P2x:"))
P2y = int(input("Enter with P2y:"))
```

```python
plt.imshow(img)
```

Negative filter 

```python
img[P1x:P2x,P1y:P2y] = 255 - img[P1x:P2x,P1y:P2y]
```
