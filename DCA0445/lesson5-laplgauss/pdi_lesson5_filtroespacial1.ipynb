```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
img = Image.open('regulus.png').convert("L") 
img = np.asarray(img)
plt.imshow(img,cmap="gray")
```

Averaging Filter

```python
print(img.shape)
blur_5x5 = np.ones((5,5),np.float32)/25
img_avg = cv2.filter2D(img,-1,blur_5x5)
```

Kernel 5x5. 

1/25

```python
blur_5x5
```

```python
f, ax = plt.subplots(1, 2, figsize=(20, 15))
ax[0].imshow(img, cmap = 'gray')
ax[0].set_title('Original')
ax[1].imshow(img_avg, cmap = 'gray')
ax[1].set_title('Averaging')
plt.show()
```

Edge Detection

```python
edge_detection = np.array([[-1, -1, -1], [-1, 2, -1], [-1, -1, -1]])
```

```python
edge_detection
```

```python
img.shape
```

```python
img_edge_detection = cv2.filter2D(img,-1,edge_detection)
```

```python
f, ax = plt.subplots(1, 2, figsize=(20, 15))
ax[0].imshow(img, cmap = 'gray')
ax[0].set_title('Original')
ax[1].imshow(img_edge_detection, cmap = 'gray')
ax[1].set_title('Edge Detection')
plt.show()
```

Gaussian + Laplace filter

```python
laplace = np.array([[0, -1, 0], [-1, 4, -1], [0, -1, 0]])
gauss = np.array([[0.0625,0.125,0.0625],[0.125,0.25,0.125],[0.0625,0.125,0.0625]])
```

Gaussian

```python
img_gauss = cv2.filter2D(img,-1,gauss)
```

```python
f, ax = plt.subplots(1, 2, figsize=(20, 15))
ax[0].imshow(img, cmap = 'gray')
ax[0].set_title('Original')
ax[1].imshow(img_gauss, cmap = 'gray')
ax[1].set_title('Gaussian Filter')
plt.show()
```

Laplace

```python
img_laplace = cv2.filter2D(img,-1,laplace)
```

```python
f, ax = plt.subplots(1, 2, figsize=(20, 15))
ax[0].imshow(img, cmap = 'gray')
ax[0].set_title('Original')
ax[1].imshow(img_laplace, cmap = 'gray')
ax[1].set_title('Laplace Filter')
plt.show()
```

Laplace in Gaussian Image

```python
img_gauss_laplace = cv2.filter2D(img_gauss,-1,laplace)
```

```python
f, ax = plt.subplots(1, 2, figsize=(20, 15))
ax[0].imshow(img, cmap = 'gray')
ax[0].set_title('Original')
ax[1].imshow(img_gauss_laplace, cmap = 'gray')
ax[1].set_title('Laplace + Gauss')
plt.show()
```

Full

```python
fig, axes = plt.subplots(2, 2, figsize=(20, 15))
axes[0,0].imshow(img, cmap = 'gray')
axes[0,0].set_title('Original')

axes[0,1].imshow(img_gauss, cmap = 'gray')
axes[0,1].set_title('Gaussian Filter')

axes[1,0].imshow(img_laplace, cmap = 'gray')
axes[1,0].set_title('Laplace Filter')

axes[1,1].imshow(img_gauss_laplace, cmap = 'gray')
axes[1,1].set_title('Laplace + Gauss')
plt.show()
```
